```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/Cat_image.webp")
cv2_imshow(image)
```

```python
print("SHAPE:", image.shape)
# 630 => height
# 630 => width
# 3 => color channels (RGB) Red, Green, Blue (BGR) =>. Blue, Green, Red
print("HEIGHT:", image.shape[0])
print("WIDTH:", image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (400, 300))
cv2_imshow(resized)
```

```python
cropped = gray[100:300, 150:350]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped_cat.jpg", flipped)
```

```python
cv2.imwrite("cropped_cat.jpg", cropped)
```
